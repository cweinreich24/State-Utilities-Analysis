```python
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
file = "generationcosts.csv"
df = pd.read_csv(file)
```

```python
pprint(df.columns)
df
```

```python
df.boxplot("Levelized cost of energy", by="Gen Type", figsize=(30, 10), grid=True)
plt.ylabel("$/kWh")
plt.savefig("BoxPlot by Gen Type", dpi=200)
plt.show()
```

```python
group_BIO = df[df["Gen Type"] == "Biopower"]["Levelized cost of energy"]
group_CT = df[df["Gen Type"] == "Combustion Turbine"]["Levelized cost of energy"]
group_PV = df[df["Gen Type"] == "Photovoltaic"]["Levelized cost of energy"]
group_CC = df[df["Gen Type"] == "Combined Cycle"]["Levelized cost of energy"]
group_COW = df[df["Gen Type"] == "Coal"]["Levelized cost of energy"]
group_GEO = df[df["Gen Type"] == "Geothermal"]["Levelized cost of energy"]
group_HY = df[df["Gen Type"] == "Hydroelectric Gen Type"]["Levelized cost of energy"]
group_LBW = df[df["Gen Type"] == "Land-Based Wind"]["Levelized cost of energy"]
group_NUC = df[df["Gen Type"] == "Nuclear"]["Levelized cost of energy"]
```

```python
stats.f_oneway(group_CT, group_NUC, group_CC, group_PV)
```
